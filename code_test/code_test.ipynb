```python
import pandas as pd

# 假设这是你的数据框
data = {'column_name': ['apple', 'banana', 'apple', 'orange', 'banana', 'apple', 'kiwi']}
df = pd.DataFrame(data)

# 计算每个字符串的频数，并按频数降序排序
frequency_df = df['column_name'].value_counts().reset_index()
frequency_df.columns = ['string', 'frequency']
frequency_df = frequency_df.sort_values(by='frequency', ascending=False).reset_index(drop=True)

# 找到'target_string'所在的索引
target_string = 'ot'
target_index = frequency_df.index[frequency_df['string'] == target_string].tolist()

# 计算'target_string'之后所有字符串的频数和
if target_index:
    total_frequency_after_target = frequency_df.loc[target_index[0] + 1:, 'frequency'].sum()
else:
    total_frequency_after_target = 0

print(f"在 '{target_string}' 之后的所有字符串的频数总和是: {total_frequency_after_target}")
```
